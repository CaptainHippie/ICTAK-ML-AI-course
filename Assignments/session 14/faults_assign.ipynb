```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/faults.csv')
df.head()
```

```python
df.duplicated().any()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
data = df.values
X = data[:, :-1]
y = data[:, -1]
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=1)
```

```python
mean_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
median_imputer = SimpleImputer(missing_values=np.nan, strategy='median')
mode_imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

```python
X_train[:, (15,)] = mean_imputer.fit_transform(X_train[:, (15,)])
X_test[:, (15,)] = mean_imputer.transform(X_test[:, (15,)])

X_train[:, (1,)] = median_imputer.fit_transform(X_train[:, (1,)])
X_test[:, (1,)] = median_imputer.transform(X_test[:, (1,)])

X_train[:, (13,)] = mode_imputer.fit_transform(X_train[:, (13,)])
X_test[:, (13,)] = mode_imputer.transform(X_test[:, (13,)])
```

```python
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
min_max_scaler = MinMaxScaler(feature_range=(0,1))
```

```python
X_train = min_max_scaler.fit_transform(X_train)
X_test = min_max_scaler.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

Naive Bayes Classification

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
y_pred_gnb = gnb.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_gnb)
```

KNN Classification

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
y_pred_knn = knn.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_knn)
```

SVM Classification

```python
clf = SVC()
clf.fit(X_train, y_train)
```

```python
y_pred_svm = clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_svm)
```

```python
precision_score(y_test, y_pred_svm, average='weighted')
```

```python
recall_score(y_test, y_pred_svm, average='weighted')
```
