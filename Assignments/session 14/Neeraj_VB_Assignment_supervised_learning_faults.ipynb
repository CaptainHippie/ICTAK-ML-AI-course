```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/faults.csv')
df.head(20)
```

Replacing zeros with null

```python
df.duplicated().any()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
data = df.values
X = data[:, :-1]
y = data[:, -1]
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
```

Mean - Empty_Index, (column:15)

Mode - Steel_Plate_Thickness, (column:13)

Median - X_Maximum (column:1)

```python
mean_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
median_imputer = SimpleImputer(missing_values=np.nan, strategy='median')
mode_imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
min_max_scaler = MinMaxScaler(feature_range=(0,1))
```

```python
X_train[:, (15,)] = mean_imputer.fit_transform(X_train[:, (15,)])
X_train[:, (1,)] = median_imputer.fit_transform(X_train[:, (1,)])
X_train[:, (13,)] = mode_imputer.fit_transform(X_train[:, (13,)])

X_test[:, (15,)] = mean_imputer.transform(X_test[:, (15,)])
X_test[:, (1,)] = median_imputer.transform(X_test[:, (1,)])
X_test[:, (13,)] = mode_imputer.transform(X_test[:, (13,)])
```

```python
pd.DataFrame(X_train).isna().any()
```

```python
pd.DataFrame(X_train)
```

Applying Label encoding on target column

```python
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

Applying min-max scaling on all independant variables

```python
X_train = min_max_scaler.fit_transform(X_train)
X_test = min_max_scaler.transform(X_test)
```

```python
pd.DataFrame(X_train)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score, f1_score
```

## Naive Bayes Classification

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
y_pred_gnb = gnb.predict(X_test)
```

```python
accuracy_gnb = accuracy_score(y_test, y_pred_gnb)
precision_gnb = precision_score(y_test, y_pred_gnb, average='weighted')
recall_gnb = recall_score(y_test, y_pred_gnb, average='weighted')
f1_gnb = f1_score(y_test, y_pred_gnb, average='weighted')
```

```python
cr_gnb = classification_report(y_test, y_pred_gnb, output_dict=True, zero_division='warn')
pd.DataFrame(cr_gnb)
```

## KNN Classification

##### Selecting best value for n_neighbors using GridsearchCV

```python
knn_test = KNeighborsClassifier()
#create a dictionary of all values we want to test for n_neighbors
param_grid = {'n_neighbors': np.arange(1, 40)}
#use gridsearch to test all values for n_neighbors
knn_gscv = GridSearchCV(knn_test, param_grid, cv=5)
knn_gscv.fit(X_train, y_train)
```

```python
knn_gscv.best_params_
```

Here, 1 gives best result for n_neighbors

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
y_pred_knn = knn.predict(X_test)
```

```python
accuracy_knn = accuracy_score(y_test, y_pred_knn)
precision_knn = precision_score(y_test, y_pred_knn, average='weighted')
recall_knn = recall_score(y_test, y_pred_knn, average='weighted')
f1_knn = f1_score(y_test, y_pred_knn, average='weighted')
```

```python
cr_knn = classification_report(y_test, y_pred_knn, output_dict=True, zero_division='warn')
pd.DataFrame(cr_knn)
```

## SVM Classification

```python
svm = SVC()
svm.fit(X_train, y_train)
```

```python
y_pred_svm = svm.predict(X_test)
```

```python
accuracy_svm = accuracy_score(y_test, y_pred_svm)
precision_svm = precision_score(y_test, y_pred_svm, average='weighted')
recall_svm = recall_score(y_test, y_pred_svm, average='weighted')
f1_svm = f1_score(y_test, y_pred_svm, average='weighted')
```

```python
cr_svm = classification_report(y_test, y_pred_svm, output_dict=True, zero_division='warn')
pd.DataFrame(cr_svm)
```

## Comparing Classification Scores

```python
comparison_data = (('Naive Bayes', accuracy_gnb, precision_gnb, recall_gnb, f1_gnb),
                ('K Nearest Neighbor', accuracy_knn, precision_knn, recall_knn, f1_knn),
                ('Support Vector', accuracy_svm, precision_svm, recall_svm, f1_svm))
```

```python
pd.DataFrame(comparison_data, columns = ('Comparison algorithm','Accuracy','Precision','Recall','F-measure'))
```
