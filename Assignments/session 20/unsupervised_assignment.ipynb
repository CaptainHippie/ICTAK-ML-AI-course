```python
df = pd.read_csv('USArrests.csv')
data = df.rename(columns={'Unnamed: 0':'State'})
data
```

```python
plt.figure(figsize=(15,4))
data.groupby('State')['Murder'].max().plot(kind='bar')
```

```python
plt.figure(figsize=(15,4))
data.groupby('State')['Assault'].max().plot(kind='bar')
```

```python
plt.figure(figsize=(15,4))
data.groupby('State')['Rape'].max().plot(kind='bar')
```

```python
plt.figure(figsize=(15,4))
data.groupby('State')['UrbanPop'].max().plot(kind='bar')
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dat = data.iloc[:,1:].values
```

```python
plt.figure(figsize=(20,8))
dendrogram = sch.dendrogram(sch.linkage(dat, method  = "ward"))
plt.title('Dendrogram')
plt.xlabel('Crime Rate')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
AC = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage ='ward')
```

```python
agg_fit =AC.fit(dat)
agg_fit.labels_
```

```python
data['cluster labels']= agg_fit.labels_
```

```python
data[['State','cluster labels']]
```

```python
Safe= data.groupby('cluster labels')['State'].unique()[0]
Safe
```

```python
Moderate= data.groupby('cluster labels')['State'].unique()[2]
Moderate
```

```python
Danger= data.groupby('cluster labels')['State'].unique()[1]
Danger
```

```python
plt.figure(figsize=(10,5))
plt.scatter(dat[agg_fit.labels_==0, 0], dat[agg_fit.labels_==0, 1], s=100, c='yellow', label ='Safe_Zone')
plt.scatter(dat[agg_fit.labels_==1, 0], dat[agg_fit.labels_==1, 1], s=100, c='red', label ='Danger_Zone')
plt.scatter(dat[agg_fit.labels_==2, 0], dat[agg_fit.labels_==2, 1], s=100, c='orange', label ='Moderate_Zone')
plt.legend()
plt.show()
```
