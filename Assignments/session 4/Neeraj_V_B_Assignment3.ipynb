```python
import pandas as pd

data = pd.read_csv('data.csv')
```

```python
data
```

checking for null values

```python
data.isna().sum()
```

filling missing values in calories column with its mean value

```python
data['Calories'].fillna(data['Calories'].mean(), inplace=True)
data
```

converting the Date column to pandas timestamp format

```python
data['Date']= pd.to_datetime(data['Date'])
data
```

Replacing missing dates with an approximate date if its in sequence.
Also replacing out of range duration by checking for typo in third digit or else replace with max value, that is 60

```python
for index, row in data.iterrows():
    if pd.isna(row.Date):
        if (data.at[index - 1, 'Date'] + pd.Timedelta(days=2)) == data.at[index + 1, 'Date']:
            data.at[index, 'Date'] = data.at[index - 1, 'Date'] + pd.Timedelta(days=1)
        
    if (data.at[index, 'Duration'] > 60) or (data.at[index, 'Duration'] < 30):
        if row.Duration > 100:
            truncated = int(row.Duration / 10)
            if (truncated >= 30) and (truncated <= 60):
                data.at[index, 'Duration'] = truncated
        else:
            data.at[index, 'Duration'] = 60
```

```python
data
```

Writing cleaned data to a csv file

```python
data.to_csv('cleaned_data.csv')
```
