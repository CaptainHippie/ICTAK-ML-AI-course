##### extract the features using tfidf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
features
```

```python
tfidf = TfidfVectorizer(stop_words= "english")
```

```python
tfidf_matrix = tfidf.fit_transform(features)
```

```python
tfidf_matrix.shape
```

```python
#### cosine similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
similarity.shape
```

```python
similarity[4]
```

```python
def restaurant_rec_sys(name, similarity = similarity):
    index = indices[name]
    similarity_scores = list(enumerate(similarity[index]))
    similarity_scores = sorted(similarity_scores, key = lambda x: x[1],reverse = True)
    similarity_scores = similarity_scores[:11]
    res_indices = [i[0] for i in similarity_scores]
    return df2['Name'].iloc[res_indices]
```

```python
restaurant_rec_sys("The Clam Bar")
```
