```python
bill_amt_features = ["bill_amt" + str(i) for i in range(1,7)]
pay_amt_features = ["pay_amt" + str(i) for i in range(1,7)]
```

```python
# pay_amt_features
```

```python
binary_features = ['male', 'grad_school', 'university', 'married']
```

```python
pay_features = ["pay_" + str(i) for i in range(1,7)]
```

```python
 num_features = ["limit_bal", "age"] + bill_amt_features + pay_amt_features + pay_features
```

```python
x = data[num_features + binary_features]
```

```python
y = data['default']
```

```python
pd.DataFrame(x)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,
                                                    test_size=0.2, 
                                                    random_state = 134)
```

```python
# x_train
```

```python
sc_features = ["limit_bal", "age"] + bill_amt_features + pay_amt_features
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train.loc[:,sc_features] = sc.fit_transform(x_train[sc_features])
```

```python
x_test.loc[:,sc_features] = sc.fit_transform(x_test[sc_features])
```

```python
from sklearn.cluster import KMeans
from sklearn.ensemble import GradientBoostingClassifier
```

```python
kmc = KMeans(n_clusters=3)
```

```python
kmc.fit(x_train,y_train)
```

```python
y_pred_kmc = kmc.predict(x_test)
y_pred_kmc
```

```python
from sklearn.metrics import accuracy_score, precision_score
```

```python
accuracy_kmc = accuracy_score(y_test, y_pred_kmc)
accuracy_kmc
```

```python
gb = GradientBoostingClassifier(learning_rate=0.1)
gb.fit(x_train,y_train)
```

```python
y_pred_grad = gb.predict(x_test)
y_pred_grad
```

```python
accuracy_grad = accuracy_score(y_test, y_pred_grad)
accuracy_grad
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model=LogisticRegression()
log_model.fit(x_train, y_train)
```

```python
y_pred_log = log_model.predict(x_test)
y_pred_log
```

```python
accuracy_log = accuracy_score(y_test, y_pred_log)
accuracy_log
```
