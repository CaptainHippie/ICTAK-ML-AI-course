```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.array([23,34,56])
y = np.array([120,345,567])
```

```python
plt.plot(x,y)
```

```python
x2 = np.array([12,28,34,56,82,96])
y2 = np.array([120,345,567,345,238,856])
```

```python
plt.plot(x2,y2,'o')
```

```python
z = np.array([120,345,567,345,238,856])
plt.plot(z, 'o:g', marker='^')
```

o -circle
* -star
. -point
x -X

+ -plus
P -plus filled
s -Square
D -Diamond
d -thin diamond
p -pentagon
H/h -hexagon
v -triangle down
^ -triangular up
< -triangle left
> -triangle right
1 -tri down

4 -tri right
| -vline

```python
font1={'family':'comic sans ms','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}

plt.plot(x,y,x2,y2)
plt.title("Some Graph", fontdict=font1, loc='left')
plt.xlabel('days', fontdict=font2)
plt.ylabel('scores', fontdict=font2)
plt.grid(axis='y', color='blue', linestyle='--',linewidth=0.5)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(x2,y2)
```

```python
plt.scatter(x,y, color='red', alpha=0.3)
plt.scatter(x2,y2, color='darkviolet')
```

```python
plt.bar(x2,y2, width=4)
```
