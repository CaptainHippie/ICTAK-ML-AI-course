```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
data = df.values
x = data[:, :-1]
y = data[:, -1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
le = LabelEncoder()
min_max = MinMaxScaler(feature_range=(0,1))
```

```python
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
x_train = min_max.fit_transform(x_train)
x_test = min_max.transform(x_test)
```

```python
pd.DataFrame(x_train).head(10)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(x_train, y_train)
```

```python
y_pred = gnb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, classification_report
```

```python
accuracy_score(y_pred, y_test)
```

```python
pd.DataFrame(classification_report(y_pred, y_test,  output_dict=True, zero_division='warn'))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(x_train, y_train)
```

```python
y_pred_log = log.predict(x_test)
```

```python
accuracy_score(y_pred_log, y_test)
```

```python
pd.DataFrame(classification_report(y_pred_log, y_test,  output_dict=True, zero_division='warn'))
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(x_train, y_train)
```

```python
y_pred_lda = lda.predict(x_test)
```

```python
accuracy_score(y_pred_lda, y_test)
```

```python
pd.DataFrame(classification_report(y_pred_lda, y_test,  output_dict=True, zero_division='warn'))
```
