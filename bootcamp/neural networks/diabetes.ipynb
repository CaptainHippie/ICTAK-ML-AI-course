```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('diabetes.csv')
df.head(10)
```

```python
df.info()
```

```python
df['Glucose'].replace(0, np.nan, inplace=True)
df['BloodPressure'].replace(0, np.nan, inplace=True)
df['SkinThickness'].replace(0, np.nan, inplace=True)
df['Insulin'].replace(0, np.nan, inplace=True)
df['BMI'].replace(0, np.nan, inplace=True)
df['DiabetesPedigreeFunction'].replace(0, np.nan, inplace=True)
df['Age'].replace(0, np.nan, inplace=True)
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
data = df.values
X = data[:, :-1]
y = data[:, -1]
pd.DataFrame(X)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=1)
```

```python
mean_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
median_imputer = SimpleImputer(missing_values=np.nan, strategy='median')
```

```python
X_train[:, (1,2,3,5)] = mean_imputer.fit_transform(X_train[:, (1,2,3,5)])
X_train[:, (4,6,7)] = median_imputer.fit_transform(X_train[:, (4,6,7)])
X_test[:, (1,2,3,5)] = mean_imputer.transform(X_test[:, (1,2,3,5)])
X_test[:, (4,6,7)] = median_imputer.transform(X_test[:, (4,6,7)])
```

```python
pd.DataFrame(X_train).isna().any()
```

```python
pd.DataFrame(X_train).head(10)
```

```python
X_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model1 = Sequential()
model1.add(Dense(128, input_shape = (8,), activation = "relu"))
model1.add(Dense(64, activation = "relu"))
model1.add(Dense(64, activation = "sigmoid"))
model1.add(Dense(32, activation = "sigmoid"))
model1.add(Dense(32, activation = "sigmoid"))
model1.add(Dense(1, activation = "sigmoid"))
model1.summary()
```

```python
model1.compile(optimizer='adam', loss = "mean_squared_error", metrics = ["accuracy"])
model1.fit(X_train, y_train, epochs=20, batch_size=128)
```

```python
model1.evaluate(X_test, y_test)[1]
```

```python
y_pred = (model1.predict(X_test) > 0.5).astype('int32')
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
