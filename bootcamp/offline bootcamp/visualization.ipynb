```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head(20)
```

```python
sns.displot(tips, x='total_bill', col='time', kde=True)
```

```python
sns.boxplot(data=tips, x='day', y='total_bill')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.lineplot(data=tips, x='day', y='tip', hue='sex')
```

```python
sns.heatmap(tips[['total_bill','tip','size']])
```

```python
sns.stripplot(tips)
```

white dot in violinplot is the median

```python
sns.violinplot(tips, x='total_bill', y='day')
```
